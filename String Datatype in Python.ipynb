# String
A String is a data structure in Python that represents a sequence of characters. It is an immutable data type, meaning that once you have created a string, you cannot change it. Strings are used widely in many different applications, such as storing and manipulating text data, representing names, addresses, and other types of data that can be represented as text.

### Creating String
* Strings are charachters which are enclosed in single quotes (‘ ‘), double quotes (” “), and triple double quotes (“”” “””).
* The triple quotes can be used to declare multiline strings multiline comments and as doc strings

```python
# creating a String
s1= 'this is a string enclosed in a single quote'
s2= "this is a string enclosed in a double quotes"
s3= '''this is a multistring enclosed in triple-single quotes'''
s4= """this is a multistring enclosed in triple-double quotes"""
print(s1,s2,s3,s4, sep='\n')
```

### Escape Sequencing
* While printing Strings with single and double quotes in it causes SyntaxError because String already contains Single and Double Quotes and hence cannot be printed with the use of either of these. Hence, to print such a String either Triple Quotes are used or Escape sequences are used to print Strings. 
* Escape sequences start with a backslash and can be interpreted differently. If single quotes are used to represent a string,  then all the single quotes present in the string must be escaped and the same is done for Double Quotes. 

```python
a="this cup now belong's to INDIA"  # syntax error  for single quotes because of belongs has single quote in it
a='this cup now belong\'s to INDIA' 
```

```python
print("hello\'world'") #single quote
print("hello\\world") #insert single backslash
print("hello\nworld") #new line
print("hello\rworld") # takes the cursor to the first position of the same line
print("hello\tworld") #tab add 4 spaces 
print("hello\bworld") # takes cursor one position backward
print("hello\fworld") #form feed is page breaking
print("hello\ooworld") #octal value
```

### Raw String 
* In Python, a raw string is a special type of string that allows you to include backslashes (\) without interpreting them as escape sequences. 
* Raw strings are particularly useful when working with regular expressions, as they allow you to specify patterns that may contain backslashes without having to escape them
* To convert the regular string to raw string ,we use 'r' before string declaration


```python
print('c:\abc\newfolder\j.jpg') #system is not taking first letter
```

```python
print('c:\\abc\\newfolder\\j.jpg')
```

```python
print(r'c:\abc\newfolder\j.jpg') #'r'is raw string
```

### Formating string or f-String
* String formatting in Python allows you to create dynamic strings by combining variables and values.
* To create an f-string in Python, prefix the string with the letter “f”. The string itself can be formatted in much the same    way that you would with str. format(). 
* F-strings provide a concise and convenient way to embed Python expressions inside string  literals for formatting


```python
name = 'Ayesha'
print(f"My name is {name}.") #'f' is f-string
```

### Indexing
* Indexing in Python means referring to an element of an iterable by its position within the iterable.
* Each character can be accessed using their index number.
* To access characters in a string we have two ways:
1. Positive index number
2. Negative index number

##### Positive index number
* in posityive indexing,we move from left to right and  start index is '0' and the end index is 'n-1',where 'n' is 'len of the string'
##### Negative index number
* in negative indexing, we move from right to left start index is '-1' and the end index is '-n', where 'n' is 'len of the string'


```python
X = 'Python'
print(X[3])# accessing h (using postive index)
```

```python
print(X[-3])# using negative index
```

### slicing  
* Slicing is a way to extract a part of a sequence, such as a string, list, or tuple, in Python.
* You can slice a sequence by specifying the start and end indices separated by a colon [:].
* The syntax for slicing is as follows: 
syn:[start_index:end_index]


```python
S='ABCDEFGHI'
print(len(S)) #start with 0 and end with n+1
print(S[0:5])
print(S[0:8])
print(S[0:8:2]) #step 2 skipping 1 place
print(S[0:9:1]) #step 1 skipping 0 place
print(S[::]) 
print(S[4:8]) 
```

##### Reverse indexing
* Slicing in Python is also used for reversing the string.
* Slicing is the most efficient way to reverse the string compared to other reversing techniques in Python.
* Programmers usually use a slice operator as it takes less time to reverse the string. 


```python
x='python'
print(x)
print(x[::-1]) # reverse indexing
print(x[-1:-7:-1]) # '-' says that right to left
```

```python
print(x[6:0:-1])
```
